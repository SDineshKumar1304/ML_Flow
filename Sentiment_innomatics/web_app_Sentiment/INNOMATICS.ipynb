# **Text Cleaning**

```python
import nltk
nltk.download('punkt')
```

```python
nltk.download('wordnet')
```

```python
nltk.download('stopwords')
```

```python
import pandas as pd
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from bs4 import BeautifulSoup
import unicodedata
from nltk.stem import WordNetLemmatizer

def clean_text(sentence):
    if not isinstance(sentence, str):
        return ""
    text = sentence.lower()

    text = BeautifulSoup(text, 'html.parser').get_text()

    text = ' '.join([word for word in text.split() if not word.startswith('http')])

    text = ''.join([char for char in text if char not in string.punctuation + '’‘'])

    text = ''.join([i for i in text if not i.isdigit()])

    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    text = ' '.join([word for word in word_tokens if word.lower() not in stop_words])

    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')

    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    text = ' '.join([lemmatizer.lemmatize(word) for word in tokens])

    return text
```

```python
data_bat['review_text']=data_bat['review text'].apply(clean_text)
```

# **Model Training**

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    data_bat['text'], data_bat['sentiment'], test_size=0.2, random_state=42
)
```

# **Setting up the Pipeline and Performing Sampling**

```python
pipeline = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('sampler', SMOTE(sampling_strategy=0.5)),
    ('classifier', MultinomialNB())
])
```

```python

pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score
y_pred = pipeline.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

# **Fine Tuning the Model**

```python
from sklearn.model_selection import GridSearchCV

pipeline = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('sampler', SMOTE(sampling_strategy=0.5)),
    ('classifier', MultinomialNB())
])

param_grid = {
    'vectorizer__ngram_range': [(1, 1), (1, 2)],
    'vectorizer__min_df': [1, 2, 3],
    'classifier__alpha': [0.1, 0.5, 1.0],
}
```

```python
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='f1_macro')
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
print("Best Parameters:", best_params)
```

# **Tuned Model Evaluation**

```python
best_model = grid_search.best_estimator_
y_pred_tuned = best_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print("Tuned Model Accuracy:", accuracy_score(y_test, y_pred_tuned))
print("Tuned Model Classification Report:\n", classification_report(y_test, y_pred_tuned))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
conf_matrix = confusion_matrix(y_test, y_pred_tuned, labels=['negative', 'positive'])

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['negative', 'positive'], yticklabels=['negative', 'positive'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
